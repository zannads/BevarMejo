```python
import os
#add ../pybeme to the path
import sys
sys.path.append('../pybeme')
sys.path.append('..')

from pybeme.beme_experiment import load_experiment_results, extract_dataframe

hanoi = load_experiment_results(os.path.join(os.curdir, '..', 'data',
                         'bemeopt__nsga2__hanoi_fbiobj', 'output', 'bemeopt__nsga2__hanoi_fbiobj__exp.json') )
```

```python
import numpy as np
import plotly.graph_objects as go

# Extract the fitness vector of the final solution from the 'mixed' dictionary
fig = go.Figure()

# for each island in islands and individual in individual extract the fitness-vector and put it in a numpy array
fitness = np.array([ind['fitness-vector'] for island in hanoi['archipelago']['islands'] for ind in island['generations'][-1]['individuals']])

# plot in the 2d space the best pareto front and limit the y axis between -1 and 0
# different colors for each experiment

fig.add_trace(go.Scatter(x=fitness[:,0], y=-fitness[:,1], mode='markers', marker=dict(size=5), name=hanoi['name']))

fig.update_layout(title='Pareto front of the final generation of the rehab and mixed experiments',
                    xaxis_title='Cost',
                    yaxis_title='Reliability Index (Todini 2000)',
                    yaxis_range=[0, 1])
fig.show()
```
